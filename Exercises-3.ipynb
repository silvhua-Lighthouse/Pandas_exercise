### How many leading (n=1) roles were available to actors, and how many to actresses, in each year of the 1950s?

```python
filter = (cast['year'] >=1950) & (cast['year'] <1960) & (cast['n'] == 1)
cast[filter].groupby(['year','type']).agg(['count'])['title']
```

```python
cast[filter].groupby(['year','type']).count()
```

### In the 1950s taken as a whole, how many total roles were available to actors, and how many to actresses, for each "n" number 1 through 5?

```python
filter = cast['year'].between(1950,1959)
cast[filter].groupby(['year','type','n']).count()['title'].unstack()
```

```python
cast[filter]['n'].value_counts()
```

### Use groupby() to determine how many roles are listed for each of the Pink Panther movies.

```python
filter = cast['title'].str.contains('Pink Panther')
cast[filter].groupby('title').agg({'title': 'count'})
```

### List, in order by year, each of the films in which Frank Oz has played more than 1 role.

```python
filter = cast['name'] == 'Frank Oz'
cast[filter]
```

```python
cast[filter].sort_values('year').groupby(['year','title','character']).agg({'n': 'count'})
```

```python
# Just for fun: List out all movies and character roles

cast[filter].sort_values(by='title')
```
